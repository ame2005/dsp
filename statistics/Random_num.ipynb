```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot

num = np.random.random(1000)
pmf = thinkstats2.Pmf(num)
```

```python
width = 0.4 / 16
thinkplot.Pmf(pmf, align='right', width=width)
thinkplot.Config(xlabel='Random num', ylabel='PMF')
```

```python
 cdf = thinkstats2.Cdf(num)
```

```python
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='random num', ylabel='CDF')
```
